```
data = pd.read_csv('/content/gdrive/My Drive/high_diamond_ranked_10min.csv')
data = pd.DataFrame(data)

print (data)
```

데이터 전처리

```
data.drop(['gameId'],axis =1 ,inplace =True)
```

```
train , test = train_test_split(data,test_size =0.3, random_state =0 )

train_y = train['blueWins']
test_y = test['blueWins']

```

```
train.drop(['blueWins'],axis =1 ,inplace =True)
test.drop(['blueWins'],axis =1 ,inplace =True)
```

```
game = np.array([[14,3,0,4,9,1,1,0,1,0,14979,6.6,17213,210,52,-3414,-1141,21.0,1497.9,20,3,1,9,4,11,0,0,0,0,18393,7.2,18354,229,51,3414,1141,22.9,1839.3]])
```

```
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(train, train_y)
print ("train accuracy: ",model.score(train,train_y))
print ("test accuracy: ",model.score(test,test_y))
```

```
print(model.predict(game))
print(model.predict_proba(game))
```
